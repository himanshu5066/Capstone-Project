```python
#  Import libraries
import pandas as pd
import numpy as np
from scipy.stats import skew


#  Feature Engineering

## A. Total square footage (example: combining multiple areas if present)
if set(['LotArea', 'GrLivArea']).issubset(df.columns):
    df['TotalSqFeet'] = df['LotArea'] + df['GrLivArea']

# Age of the house (example: using YearBuilt)
if 'YearBuilt' in df.columns:
    df['HouseAge'] = 2025 - df['YearBuilt']

##  Total bathrooms (example)
if set(['FullBath', 'HalfBath']).issubset(df.columns):
    df['TotalBathrooms'] = df['FullBath'] + 0.5 * df['HalfBath']

##  Interaction Feature (Garage * Living Area)
if set(['GarageCars', 'GrLivArea']).issubset(df.columns):
    df['Garage_x_Living'] = df['GarageCars'] * df['GrLivArea']

##  Flag for house having a garage
if 'GarageType' in df.columns:
    df['HasGarage'] = df['GarageType'].notnull().astype(int)

##  Binning â€“ LotArea into bins (optional)
if 'LotArea' in df.columns:
    df['LotAreaBin'] = pd.qcut(df['LotArea'], q=4, labels=['Small', 'Medium', 'Large', 'VeryLarge'])

# Encode LotAreaBin if created
if 'LotAreaBin' in df.columns:
    df = pd.get_dummies(df, columns=['LotAreaBin'], drop_first=True)

##  Log Transformation for skewed numerical features
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()

# Remove target if present to avoid transforming it
target_col = 'SalePrice' if 'SalePrice' in df.columns else None
if target_col:
    numeric_cols.remove(target_col)

# Check skewness
skewed_feats = df[numeric_cols].apply(lambda x: skew(x.dropna())).sort_values(ascending=False)
high_skew = skewed_feats[skewed_feats > 0.75]
skewed_cols = high_skew.index.tolist()

# Apply log1p (log(x+1)) to reduce skewness
df[skewed_cols] = np.log1p(df[skewed_cols])

#  Final check
print("Final dataset shape after feature engineering:", df.shape)
print("Sample of engineered features:\n", df.head())
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import (
    LabelEncoder, OneHotEncoder, OrdinalEncoder, 
    StandardScaler, MinMaxScaler, RobustScaler, QuantileTransformer
)
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
import warnings
warnings.filterwarnings('ignore')

print("=== REAL ESTATE DATA ENCODING TECHNIQUES DEMONSTRATION ===\n")
print(f"Dataset shape: {df.shape}")
print(f"Columns: {df.columns.tolist()}")


# Identify different types of columns
numeric_columns = df.select_dtypes(include=[np.number]).columns.tolist()
categorical_columns = df.select_dtypes(include=['object']).columns.tolist()

# Remove target variable from features
if 'PropPrice' in numeric_columns:
    numeric_columns.remove('PropPrice')

print(f"\nNumeric columns ({len(numeric_columns)}): {numeric_columns[:10]}...")
print(f"Categorical columns ({len(categorical_columns)}): {categorical_columns[:10]}...")

# Create a copy for encoding
df_encoded = df.copy()

print("\n" + "="*60)
print("1. CATEGORICAL ENCODING TECHNIQUES")
print("="*60)

# 1. LABEL ENCODING
print("\n--- 1.1 Label Encoding ---")
print("Best for: Ordinal categorical variables with natural order")

# Define ordinal mappings for quality variables
quality_mappings = {
    'Ex': 5, 'Gd': 4, 'TA': 3, 'Fa': 2, 'Po': 1, 'NA': 0
}

ordinal_columns = ['ExterQual', 'ExterCond', 'KitchenQual', 'QualFireplace']

for col in ordinal_columns:
    if col in df_encoded.columns:
        # Manual ordinal encoding for quality variables
        df_encoded[f'{col}_Ordinal'] = df_encoded[col].map(quality_mappings).fillna(0)
        print(f"{col}: {df_encoded[col].value_counts().head()}")
        print(f"Encoded as: {df_encoded[f'{col}_Ordinal'].value_counts().head()}")

# 2. ONE-HOT ENCODING
print("\n--- 1.2 One-Hot Encoding ---")
print("Best for: Nominal categorical variables with no inherent order")

# Select columns for one-hot encoding (low cardinality)
onehot_columns = []
for col in categorical_columns:
    if col in df_encoded.columns:
        unique_count = df_encoded[col].nunique()
        if unique_count <= 10 and unique_count > 2:  # Reasonable cardinality
            onehot_columns.append(col)

print(f"Columns selected for One-Hot Encoding: {onehot_columns[:5]}")

# Apply one-hot encoding
for col in onehot_columns[:3]:  # Limit to first 3 for demonstration
    if col in df_encoded.columns:
        dummies = pd.get_dummies(df_encoded[col], prefix=col, drop_first=True)
        df_encoded = pd.concat([df_encoded, dummies], axis=1)
        print(f"{col}: Created {len(dummies.columns)} dummy variables")

# 3. TARGET ENCODING (Mean Encoding)
print("\n--- 1.3 Target Encoding ---")
print("Best for: High cardinality categorical variables")

def target_encode(df, categorical_col, target_col, smoothing=1):
    """Target encoding with smoothing to prevent overfitting"""
    # Calculate global mean
    global_mean = df[target_col].mean()
    
    # Calculate category means and counts
    category_stats = df.groupby(categorical_col)[target_col].agg(['mean', 'count']).reset_index()
    category_stats.columns = [categorical_col, 'mean', 'count']
    
    # Apply smoothing
    category_stats['smoothed_mean'] = (
        (category_stats['mean'] * category_stats['count'] + global_mean * smoothing) /
        (category_stats['count'] + smoothing)
    )
    
    # Create mapping dictionary
    mapping = dict(zip(category_stats[categorical_col], category_stats['smoothed_mean']))
    
    return df[categorical_col].map(mapping).fillna(global_mean)

# Apply target encoding to high cardinality categorical columns
high_cardinality_cols = ['Neighborhood', 'Street']
for col in high_cardinality_cols:
    if col in df_encoded.columns:
        df_encoded[f'{col}_TargetEncoded'] = target_encode(df_encoded, col, 'PropPrice')
        print(f"{col}: Target encoded (unique values: {df_encoded[col].nunique()})")

print("\n" + "="*60)
print("2. NUMERICAL ENCODING TECHNIQUES")
print("="*60)

# 2.1 STANDARDIZATION (Z-Score Normalization)
print("\n--- 2.1 Standardization (Z-Score) ---")
print("Best for: Features with different scales, normal distribution")

scaler_standard = StandardScaler()
numerical_features = ['PropertySize', 'GrLivArea', 'YearBuilt', 'OverallQual']

for col in numerical_features:
    if col in df_encoded.columns:
        df_encoded[f'{col}_Standardized'] = scaler_standard.fit_transform(df_encoded[[col]])
        print(f"{col}: Mean={df_encoded[col].mean():.2f}, Std={df_encoded[col].std():.2f}")
        print(f"  After standardization: Mean={df_encoded[f'{col}_Standardized'].mean():.2f}, Std={df_encoded[f'{col}_Standardized'].std():.2f}")

# 2.2 MIN-MAX NORMALIZATION
print("\n--- 2.2 Min-Max Normalization ---")
print("Best for: Features with known bounds, preserving relationships")

scaler_minmax = MinMaxScaler()
for col in numerical_features[:2]:  # Limit for demonstration
    if col in df_encoded.columns:
        df_encoded[f'{col}_MinMax'] = scaler_minmax.fit_transform(df_encoded[[col]])
        print(f"{col}: Min={df_encoded[col].min():.2f}, Max={df_encoded[col].max():.2f}")
        print(f"  After normalization: Min={df_encoded[f'{col}_MinMax'].min():.2f}, Max={df_encoded[f'{col}_MinMax'].max():.2f}")

# 2.3 ROBUST SCALING
print("\n--- 2.3 Robust Scaling ---")
print("Best for: Features with outliers")

scaler_robust = RobustScaler()
outlier_prone_cols = ['PropertySize', 'GrLivArea']
for col in outlier_prone_cols:
    if col in df_encoded.columns:
        df_encoded[f'{col}_Robust'] = scaler_robust.fit_transform(df_encoded[[col]])
        print(f"{col}: Median={df_encoded[col].median():.2f}, IQR={df_encoded[col].quantile(0.75) - df_encoded[col].quantile(0.25):.2f}")

# 2.4 QUANTILE TRANSFORMATION
print("\n--- 2.4 Quantile Transformation ---")
print("Best for: Making features more Gaussian-like")

transformer_quantile = QuantileTransformer(output_distribution='normal')
for col in ['PropertySize']:
    if col in df_encoded.columns:
        df_encoded[f'{col}_Quantile'] = transformer_quantile.fit_transform(df_encoded[[col]])
        print(f"{col}: Transformed to approximately normal distribution")

print("\n" + "="*60)
print("3. ADVANCED ENCODING TECHNIQUES")
print("="*60)

# 3.1 FREQUENCY ENCODING
print("\n--- 3.1 Frequency Encoding ---")
print("Best for: Categorical variables where frequency matters")

freq_encode_cols = ['Neighborhood', 'PropertyClass']
for col in freq_encode_cols:
    if col in df_encoded.columns:
        freq_map = df_encoded[col].value_counts().to_dict()
        df_encoded[f'{col}_Frequency'] = df_encoded[col].map(freq_map)
        print(f"{col}: Frequency encoded (most common: {max(freq_map.values())} occurrences)")

# 3.2 BINARY ENCODING
print("\n--- 3.2 Binary Encoding ---")
print("Best for: High cardinality categorical variables (more efficient than one-hot)")

def binary_encode(series):
    """Simple binary encoding implementation"""
    # Get unique values and create mapping
    unique_vals = series.unique()
    n_bits = int(np.ceil(np.log2(len(unique_vals))))
    
    # Create binary representation
    binary_cols = []
    for i in range(n_bits):
        binary_cols.append(f"{series.name}_binary_{i}")
    
    # Create mapping
    mapping = {}
    for idx, val in enumerate(unique_vals):
        binary_rep = format(idx, f'0{n_bits}b')
        mapping[val] = [int(bit) for bit in binary_rep]
    
    # Apply mapping
    result_df = pd.DataFrame(index=series.index)
    for val in series:
        if val in mapping:
            for i, bit in enumerate(mapping[val]):
                col_name = f"{series.name}_binary_{i}"
                if col_name not in result_df.columns:
                    result_df[col_name] = 0
                result_df.loc[series[series == val].index, col_name] = bit
    
    return result_df

# Apply binary encoding to a high cardinality column
if 'Neighborhood' in df_encoded.columns:
    binary_encoded = binary_encode(df_encoded['Neighborhood'])
    df_encoded = pd.concat([df_encoded, binary_encoded], axis=1)
    print(f"Neighborhood: Binary encoded into {len(binary_encoded.columns)} columns")

# 3.3 POLYNOMIAL FEATURES
print("\n--- 3.3 Polynomial Features ---")
print("Best for: Capturing non-linear relationships")

from sklearn.preprocessing import PolynomialFeatures

poly_features = PolynomialFeatures(degree=2, interaction_only=True, include_bias=False)
selected_numeric = ['OverallQual', 'GrLivArea']

if all(col in df_encoded.columns for col in selected_numeric):
    poly_array = poly_features.fit_transform(df_encoded[selected_numeric])
    poly_feature_names = poly_features.get_feature_names_out(selected_numeric)
    
    for i, name in enumerate(poly_feature_names[len(selected_numeric):]):  # Skip original features
        df_encoded[f'Poly_{name}'] = poly_array[:, len(selected_numeric) + i]
    
    print(f"Created {len(poly_feature_names) - len(selected_numeric)} polynomial features")

print("\n" + "="*60)
print("4. HANDLING MISSING VALUES")
print("="*60)

# Check for missing values
missing_summary = df_encoded.isnull().sum()
missing_cols = missing_summary[missing_summary > 0]

print(f"Columns with missing values: {len(missing_cols)}")
if len(missing_cols) > 0:
    print(missing_cols.head(10))

# Different strategies for handling missing values
print("\n--- Missing Value Strategies ---")

# 1. Fill with median for numerical
numerical_cols_with_missing = []
for col in df_encoded.select_dtypes(include=[np.number]).columns:
    if df_encoded[col].isnull().sum() > 0:
        numerical_cols_with_missing.append(col)
        df_encoded[f'{col}_filled_median'] = df_encoded[col].fillna(df_encoded[col].median())

# 2. Fill with mode for categorical
categorical_cols_with_missing = []
for col in df_encoded.select_dtypes(include=['object']).columns:
    if df_encoded[col].isnull().sum() > 0:
        categorical_cols_with_missing.append(col)
        df_encoded[f'{col}_filled_mode'] = df_encoded[col].fillna(df_encoded[col].mode()[0] if not df_encoded[col].mode().empty else 'Unknown')

# 3. Create missing indicators
for col in numerical_cols_with_missing[:3]:  # Limit for demonstration
    df_encoded[f'{col}_is_missing'] = df_encoded[col].isnull().astype(int)

print(f"Applied missing value strategies to {len(numerical_cols_with_missing)} numerical and {len(categorical_cols_with_missing)} categorical columns")

print("\n" + "="*60)
print("5. FEATURE SELECTION AND MODEL COMPARISON")
print("="*60)

# Prepare different encoded datasets for comparison
datasets = {}

# Dataset 1: Basic encoding
basic_features = []
for col in df_encoded.columns:
    if col.endswith('_Ordinal') or col.endswith('_TargetEncoded') or col.endswith('_Standardized'):
        basic_features.append(col)

# Add some original numerical features
basic_features.extend(['PropertySize', 'GrLivArea', 'OverallQual', 'YearBuilt'])
basic_features = [col for col in basic_features if col in df_encoded.columns]

if len(basic_features) > 0:
    datasets['Basic Encoding'] = df_encoded[basic_features].select_dtypes(include=[np.number])

# Dataset 2: Advanced encoding
advanced_features = basic_features.copy()
for col in df_encoded.columns:
    if any(suffix in col for suffix in ['_Frequency', '_binary_', 'Poly_']):
        advanced_features.append(col)

advanced_features = [col for col in advanced_features if col in df_encoded.columns]
if len(advanced_features) > 0:
    datasets['Advanced Encoding'] = df_encoded[advanced_features].select_dtypes(include=[np.number])

# Compare model performance
print("Comparing model performance with different encoding strategies:")
print("-" * 50)

target = 'ProprtyClass'
if target in df_encoded.columns:
    y = df_encoded[target]
    
    for name, X in datasets.items():
        if len(X.columns) > 0:
            # Handle any remaining missing values
            X_clean = X.fillna(X.median())
            
            # Split data
            X_train, X_test, y_train, y_test = train_test_split(
                X_clean, y, test_size=0.2, random_state=42
            )
            
            # Train model
            rf = RandomForestRegressor(n_estimators=100, random_state=42)
            rf.fit(X_train, y_train)
            
            # Predictions
            y_pred = rf.predict(X_test)
            
            # Metrics
            mse = mean_squared_error(y_test, y_pred)
            r2 = r2_score(y_test, y_pred)
            
            print(f"{name}:")
            print(f"  Features: {len(X.columns)}")
            print(f"  RMSE: ${np.sqrt(mse):,.2f}")
            print(f"  RÂ² Score: {r2:.4f}")
            print()

print("\n" + "="*60)
print("6. ENCODING RECOMMENDATIONS SUMMARY")
print("="*60)

recommendations = """
CATEGORICAL VARIABLES:
- Low cardinality (2-10 categories): One-Hot Encoding
- Medium cardinality (10-50 categories): Target Encoding or Binary Encoding
- High cardinality (50+ categories): Target Encoding or Frequency Encoding
- Ordinal variables: Ordinal/Label Encoding with meaningful order

NUMERICAL VARIABLES:
- Normal distribution: Standardization (Z-Score)
- Bounded ranges: Min-Max Normalization
- Contains outliers: Robust Scaling
- Skewed distribution: Quantile Transformation or Log Transformation

ADVANCED TECHNIQUES:
- Non-linear relationships: Polynomial Features
- Interaction effects: Feature Crosses
- Time-based features: Cyclical Encoding for dates
- Text data: TF-IDF, Word Embeddings

MISSING VALUES:
- Numerical: Median/Mean imputation or advanced methods (KNN, iterative)
- Categorical: Mode imputation or separate 'Unknown' category
- Always consider creating missing indicators for important features
"""

print(recommendations)

print(f"\nFinal encoded dataset shape: {df_encoded.shape}")
print("Encoding pipeline completed successfully!")
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
import matplotlib.pyplot as plt
import seaborn as sns

# Step 2: Drop irrelevant columns (IDs, high-null features)
cols_to_drop = ['PropertyID', 'Alley', 'PoolQC', 'BoundaryFeatures', 'AddFeatures']
df = df.drop(columns=[col for col in cols_to_drop if col in df.columns])


# Step 3: Separate features and target
X = df.drop('PropPrice', axis=1)
y = df['PropPrice']

# Step 4: Separate numeric and categorical features
numeric_features = X.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical_features = X.select_dtypes(include=['object']).columns.tolist()

# Step 5: Preprocessing pipeline
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
])

preprocessor = ColumnTransformer(transformers=[
    ('num', numeric_transformer, numeric_features),
    ('cat', categorical_transformer, categorical_features)
])

# Step 6: Apply transformations
X_preprocessed = preprocessor.fit_transform(X)

# Step 7: Apply PCA
pca = PCA(n_components=0.95, random_state=42)  # Keep 95% variance
X_pca = pca.fit_transform(X_preprocessed)

# Step 8: Plot explained variance
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(pca.explained_variance_ratio_)+1), 
         pca.explained_variance_ratio_.cumsum(), marker='o')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Explained Variance by PCA Components')
plt.grid(True)
plt.show()

# Step 9: Final data shape
print(f'Original features: {X_preprocessed.shape[1]}')
print(f'PCA features (95% variance): {X_pca.shape[1]}')
```
